## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions:
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

## Q1. Based on the trends for the share of CVM claims as a percentage of total claims over the years 2016 through 2018, what are some business insights you can gather? What are some additional analyses you could do based on these trends?

```python
import matplotlib.pyplot as plt

# Define CVM-related (cardiac) HCPCS/CPT codes
cvm_codes = ['78452', '93000', '93010', '92950']  # Replace with the full list of cardiac procedure codes


medicare_df['is_cvm'] = medicare_df['hcpcs_code'].isin(cvm_codes)

#  Ensure only the correct years are included
filtered_medicare_df = medicare_df[medicare_df['claim_year'].isin([2016, 2017, 2018])]

# Group by year and 'is_cvm' to summarize claims
cvm_summary = (
    filtered_medicare_df
    .groupby(['claim_year', 'is_cvm'])
    .size()
    .unstack(fill_value=0)  # Creates 'CVM' and 'Non-CVM' columns
    .reset_index()
    .rename(columns={True: 'CVM', False: 'Non-CVM'})
)

#Calculate total claims for each year
cvm_summary['total_claims'] = cvm_summary['CVM'] + cvm_summary['Non-CVM']


print("CVM and Non-CVM Claims Summary (2016–2018):\n", cvm_summary)

#Plot the 100% stacked bar chart
cvm_summary[['CVM', 'Non-CVM']].div(cvm_summary['total_claims'], axis=0).plot(
    kind='bar', stacked=True, figsize=(10, 6), color=['#4C72B0', '#55A868']
)


plt.title('CVM vs Non-CVM Claims (2016–2018)')
plt.xlabel('Year')
plt.ylabel('Percentage of Claims')
plt.xticks(ticks=range(len(cvm_summary)), labels=cvm_summary['claim_year'], rotation=0)
plt.legend(['CVM Claims', 'Non-CVM Claims'], title='Claim Type')
plt.show()

```

On the other hand, non-cardiac claims remain stable and dominant, suggesting a strong and ongoing need for general healthcare services.


Low Share of CVM Claims: Cardiac-related claims make up a small part of total claims, possibly due to their specialized nature or targeted patient needs.

Gradual Increase in CVM Claims: The small rise in CVM claims suggests a growing awareness or demand for cardiac care, perhaps due to improved detection and accessibility.

Stable Demand for Non-CVM Claims: The consistent and large portion of non-cardiac claims highlights the steady need for other healthcare services, showing that cardiac care remains a more specialized area.
